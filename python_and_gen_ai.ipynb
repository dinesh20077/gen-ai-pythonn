# **COSINE SIMILARITY**


cosines similarity is a matrix used to measure how similars to vectors are ,regardless of their magnitude.

it is used in text analysis recomendation system and culestring task.



```python
import numpy as np
from numpy.linalg import norm

A=np.array([2,1,2,3,2,9])
B=np.array([3,4,2,4,5,5])

#computr cosine similarity
cosine=np.dot(A,B)/(norm(A)*norm(B))
print("cosine similarity:",cosine)
```

```python
from transformers import pipeline

summarizer= pipeline("summarization",model="facebook/bart-large-cnn")
text="""Nature is the vast, beautiful world of physical and biological phenomena,
 providing essential resources like oxygen, food, and water for all life.
  It encompasses everything from the subatomic to the cosmic, including living organisms,
  mountains, rivers, and the sky. This delicate ecosystem sustains us,
   making it our shared responsibility to protect it from harm for the well-being of future generations. """
summary=summarizer(text,max_length=30,min_length=10,do_sample=False)
print(summary[0])
```
